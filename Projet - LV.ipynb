```python
df1 = df.groupby(['Secteur'])['Nombre de 6èmes LV1 allemand', 'Nombre de 6èmes LV1 anglais', 'Nombre de 6èmes LV1 espagnol', 
                              'Nombre de 6èmes LV1 autres langues'].sum()
```

```python
df1
```

```python
df1['Sum'] = df1['Nombre de 6èmes LV1 allemand'] + df1['Nombre de 6èmes LV1 anglais'] + df1['Nombre de 6èmes LV1 espagnol'] + df1['Nombre de 6èmes LV1 autres langues']
```

```python
df1.head()
```

```python
df1['Allemand'] = round(df1['Nombre de 6èmes LV1 allemand'] / df1.Sum *100, 2)
df1['Anglais'] = round(df1['Nombre de 6èmes LV1 anglais'] / df1.Sum *100, 2)
df1['Espagnol'] = round(df1['Nombre de 6èmes LV1 espagnol'] / df1.Sum *100, 2)
df1['Autres'] = round(df1['Nombre de 6èmes LV1 autres langues'] / df1.Sum *100, 2)
```

```python
df1.reset_index()
```
